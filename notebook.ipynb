### **FEATURE ENGINEERING**

```python
# Age Groups
employee_df['Age_Group'] = pd.cut(employee_df['Age'], bins=[0, 30, 50, 100], labels=['Young', 'Middle', 'Senior'])

# Income Levels (berdasarkan quantile)
employee_df['Income_Level'] = pd.qcut(employee_df['MonthlyIncome'], q=3, labels=['Low', 'Medium', 'High'])

# Tenure Groups (untuk YearsAtCompany)
employee_df['Tenure_Group'] = pd.cut(employee_df['YearsAtCompany'], bins=[0, 2, 10, 50], labels=['New', 'Experienced', 'Veteran'])

# Promotion Frequency
employee_df['Promotion_Frequency'] = employee_df['YearsAtCompany'] / (employee_df['YearsSinceLastPromotion'] + 1)

# Tenure Ratio
employee_df['Tenure_Ratio'] = employee_df['YearsAtCompany'] / (employee_df['TotalWorkingYears'] + 1)

# Income Per Distance
employee_df['Income_Per_Distance'] = employee_df['MonthlyIncome'] / (employee_df['DistanceFromHome'] + 1)

# Tampilkan untuk verifikasi
employee_df[['Age', 'Age_Group', 'MonthlyIncome', 'Income_Level', 'YearsAtCompany', 'YearsSinceLastPromotion', 'Tenure_Group', 'Promotion_Frequency', 'Tenure_Ratio', 'Income_Per_Distance']].head()
```

## **CHI-SQUARE TEST** 
untuk menentukan apakah ada hubungan yang signifikan atau hanya kebetulan antar kolom kategorikal dengan attrition

```python
#  Chi-Square Test untuk Kolom Kategorikal vs Attrition 
from scipy.stats import chi2_contingency

print("\n\nChi-Square Test - Categorical Features vs Attrition:")
print("=" * 70)
categorical_cols_original = ['Department', 'Gender', 'MaritalStatus', 'OverTime', 'JobRole', 
                             'Age_Group', 'Income_Level', 'Tenure_Group', 'JobInvolvement', 'WorkLifeBalance', 
                             'RelationshipSatisfaction', 'JobSatisfaction', 'PerformanceRating', 
                             'EnvironmentSatisfaction', 'StockOptionLevel', 'BusinessTravel']

chi_square_results = []
for col in categorical_cols_original:
    if col in employee_df.columns:
        contingency_table = pd.crosstab(employee_df[col], employee_df['Attrition'])
        chi2, p_value, dof, expected = chi2_contingency(contingency_table)
        chi_square_results.append({
            'Feature': col,
            'Chi2_Score': chi2,
            'P_Value': p_value,
            'Significant': 'Yes' if p_value < 0.05 else 'No'
        })

chi_square_df = pd.DataFrame(chi_square_results).sort_values('Chi2_Score', ascending=False)

print("\nBerdasarkan Chi-Square Test (Categorical):")
print(chi_square_df.to_string(index=False))
```

**INSIGHTS**
1. **Kelompok "Faktor Utama" (Skor > 60)**

- `Overtime` membuktikan bahwa beban kerja adalah pemicu utama attrition.

- `JobRole` menunjukkan bahwa ada jabatan tertentu yang mungkin memiliki tingkat stres yang jauh lebih tinggi dibanding jabatan lain.

**Rekomendasi**: Perusahaan harus segera mengaudit kebijakan lembur dan memberikan perhatian khusus pada JobRole dengan beban kerja tertinggi.

2. **Kelompok "Ekonomi & Status" (Skor 35-50)**

- `StockOptionLevel` Kepemilikan saham perusahaan adalah pengikat yang sangat kuat (**skor 47.85**). Karyawan yang punya "rasa memiliki" lewat saham cenderung bertahan.

- `Income_Level` menunjukkan uang tetap jadi motivasi. Karyawan di level gaji rendah jauh lebih mudah pindah.

Selanjutnya melakukan **Uji ANOVA** untuk melihat apakah ada perbedaan gaji yang nyata dari 3 kategori `Business Travel`

```python
from scipy import stats

# Siapkan data per grup
frequent = employee_df[employee_df['BusinessTravel'] == 'Travel_Frequently']['MonthlyIncome']
rarely = employee_df[employee_df['BusinessTravel'] == 'Travel_Rarely']['MonthlyIncome']
non_travel = employee_df[employee_df['BusinessTravel'] == 'Non-Travel']['MonthlyIncome']

# One-Way ANOVA
f_stat, p_val = stats.f_oneway(frequent, rarely, non_travel)

print(f"ANOVA P-Value: {p_val}")
if p_val < 0.05:
    print("Kesimpulan: Ada perbedaan gaji yang signifikan antar kategori perjalanan.")
else:
    print("Kesimpulan: Tidak ada perbedaan gaji yang nyata (gaji cenderung sama saja).")
```

Jadi kesimpulannya, Karyawan yang sering bepergian memiliki beban kerja dan kelelahan fisik yang lebih tinggi, namun mereka TIDAK menerima kompensasi gaji yang lebih besar dibandingkan rekan mereka yang bekerja menetap di kantor.

```python
from sklearn.preprocessing import LabelEncoder


# 1. Copy dataframe agar data asli tidak berubah
employee_df_encoded = employee_df.copy()

# --- BAGIAN ENCODING ---

# A. Kolom yang diubah menjadi True/False (One-Hot Encoding)
cols_to_one_hot = ['BusinessTravel', 'JobRole', 'MaritalStatus', 'Age_Group', 'Tenure_Group']

# Gunakan pd.get_dummies untuk mengubah kolom menjadi kolom-kolom baru (0/1)
employee_df_encoded = pd.get_dummies(employee_df_encoded, columns=cols_to_one_hot, dtype=int)

# B. Kolom yang di-encode angka (Label Encoding - Tetap 1 Kolom)
if 'Income_Level' in employee_df_encoded.columns:
    le_income = LabelEncoder()
    employee_df_encoded['Income_Level'] = le_income.fit_transform(employee_df_encoded['Income_Level'].astype(str))

# C. Kolom OverTime (Manual Encode 0/1)
if 'OverTime' in employee_df_encoded.columns:
    employee_df_encoded['OverTime'] = (employee_df_encoded['OverTime'] == 'Yes').astype(int)

# D. Encode Target (Attrition)
if employee_df_encoded['Attrition'].dtype == 'object':
    le_target = LabelEncoder()
    employee_df_encoded['Attrition'] = le_target.fit_transform(employee_df_encoded['Attrition'])

# --- BAGIAN DROP KOLOM ---

features_to_drop = ['EmployeeId', 'EmployeeCount', 'StandardHours', 'Over18', 'EducationField', 'Department', 'Gender', 'Education']
existing_drop_cols = [c for c in features_to_drop if c in employee_df_encoded.columns]

# Hapus kolom yang tidak relevan dan target untuk mendefinisikan X
X = employee_df_encoded.drop(columns=['Attrition'] + existing_drop_cols)
y = employee_df_encoded['Attrition']

# --- BAGIAN FEATURE IMPORTANCE (CORRELATION) ---

corr_with_target = []
for col in X.columns:
    if pd.api.types.is_numeric_dtype(employee_df_encoded[col]):
        corr = employee_df_encoded[col].corr(employee_df_encoded['Attrition'])
        corr_with_target.append({'Feature': col, 'Correlation': abs(corr)})

feature_importance_corr = pd.DataFrame(corr_with_target).sort_values('Correlation', ascending=False)

# Plot Feature Importance (Top 40)
plt.figure(figsize=(12, 8))
sns.barplot(data=feature_importance_corr.head(40), x='Correlation', y='Feature', palette='viridis')
plt.title('Top 40 Features - Absolute Correlation with Attrition (Mixed Encoding)')
plt.show()

print(feature_importance_corr.head(40).to_string(index=False))
```

```python
employee_df_encoded.head()
```

```python
# Membagi data training dan testing
from sklearn.model_selection import train_test_split

# Gunakan fitur yang masuk dalam daftar seleksi Anda
features_selected = feature_importance_corr['Feature'].head(40).tolist()
X_final = X[features_selected]

# Membagi 80% training, 20% testing
# Gunakan stratify agar proporsi Resign (1) dan Stay (0) tetap seimbang di kedua set
X_train, X_test, y_train, y_test = train_test_split(
    X_final, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
# Mengatasi IMBALANCED Data dengan metode SMOTE
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

print(f"Jumlah data sebelum SMOTE: {len(y_train)}")
print(f"Jumlah data setelah SMOTE: {len(y_train_balanced)}")
```

```python
# 1. Cek Distribusi SEBELUM SMOTE (pada y_train)
print("Distribusi Target SEBELUM SMOTE:")
print(y_train.value_counts())
print("-" * 30)

# 2. Cek Distribusi SESUDAH SMOTE (pada y_train_balanced)
# Karena y_train_balanced biasanya berupa array numpy, kita ubah dulu ke Series pandas
y_train_resampled_series = pd.Series(y_train_balanced)

print("Distribusi Target SESUDAH SMOTE:")
print(y_train_resampled_series.value_counts())
```

```python
# Melatih Model
from sklearn.ensemble import RandomForestClassifier

# Membuat model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Melatih model dengan data yang sudah diseimbangkan
rf_model.fit(X_train_balanced, y_train_balanced)
```

## Evaluation

```python
# Evaluation Model
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

y_pred = rf_model.predict(X_test)

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

### HYPERPARAMETER TUNING

```python
from sklearn.model_selection import RandomizedSearchCV

# 1. Menentukan rentang parameter yang ingin dicoba
param_dist = {
    'n_estimators': [100, 200, 300, 500],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False],
    'class_weight': ['balanced', 'balanced_subsample', None] # Membantu menangani ketidakseimbangan kelas
}

# 2. Inisialisasi model dasar
rf = RandomForestClassifier(random_state=42)

# 3. Jalankan Pencarian Otomatis
# cv=5 artinya data dibagi 5 kali untuk validasi (Cross Validation)
# n_iter=20 artinya mencoba 20 kombinasi acak
rf_random = RandomizedSearchCV(
    estimator=rf, 
    param_distributions=param_dist, 
    n_iter=20, 
    cv=5, 
    verbose=2, 
    random_state=42, 
    n_jobs=-1,
    scoring='f1' # Kita fokus optimasi F1-Score agar seimbang antara Precision & Recall
)

# 4. Latih menggunakan data yang sudah di-SMOTE tadi
rf_random.fit(X_train_balanced, y_train_balanced)

# 5. Lihat parameter terbaik yang ditemukan
print("--- Parameter Terbaik ---")
print(rf_random.best_params_)
```

```python
# Menggunakan model terbaik dari hasil tuning
best_rf = rf_random.best_estimator_ # .best_estimator_ memanggil model dengan kombinasi parameter yang memberikan skor terbaik
y_pred_final = best_rf.predict(X_test)


print("--- EVALUASI FINAL (SETELAH TUNING) ---")
print(f"Accuracy Score: {accuracy_score(y_test, y_pred_final):.2%}")
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_final))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_final))
```

Setelah Hyperparameter Tuning Model Juga masih belum bisa memprediksi dengan baik mana Karyawan yang akan resign, itu terlihat dari nilai recall hanya `25%`.

Maka selanjutnya saya memutuskan untuk Ubah ambang batas prediksi menjadi 0.35. 

Secara default, model Machine Learning seperti Random Forest menggunakan ambang batas 0.5 (50%).

-. **Aturan Lama**: Jika peluang seseorang resign adalah 49%, model akan bilang "Stay" (0)

-. **Aturan Baru**: Karena sudah menurunkan threshold ke 0.35, seseorang dengan peluang resign 36% saja sudah langsung dicap sebagai "Resign" (1) oleh model.

```python
# Mendapatkan probabilitas (bukan langsung label 0/1)
y_prob = best_rf.predict_proba(X_test)[:, 1]

# Ubah threshold menjadi 0.35
y_pred_new = (y_prob >= 0.35).astype(int)

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_new))

print(classification_report(y_test, y_pred_new))
```

Setelah diubah ambang batas prediksi recall nya naik menjadi `44%`. Walaupun akurasi turun ke `80%`, dan resikonya akan lebih sering "salah panggil" orang (ada 23 orang yang sebenarnya setia tapi model bilang mau resign). Namun, dalam dunia HR, lebih baik salah memanggil orang setia untuk diajak bicara (retensi), daripada kehilangan orang berbakat karena model terlalu "cuek" untuk mendeteksi mereka.

Tapi sisi positifnya adalah perusahaan dapat mendeteksi lebih awal `44%` karyawan yang berisiko keluar, meningkat dari sebelumnya yang hanya `25%`.
